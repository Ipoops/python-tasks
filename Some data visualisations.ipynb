```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
donuts_per_person = {'Kwanda': 6, 'Jonathan': 8, 'Jaco': 7, 'Laura': 6, 'Andrew': 8}
```

```python
donuts = donuts_per_person.keys()
```

```python
donuts = donuts_per_person.values()
```

```python
people = donuts_per_person.keys()
```

```python
y_pos = np.arange(len(donuts_per_person))
y_pos
```

```python
plt.bar(y_pos, donuts, alpha=0.5)
plt.xticks(y_pos, people)
plt.ylabel('Donuts')
plt.title('How many donuts each')
plt.show()
```

```python
labels = donuts_per_person.keys()
sizes = donuts_per_person.values()
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'grey', 'blue']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.show()
```

```python
total_donut = {'31/01/2019': 40, '28/02/2019': 38, '31/03/2019': 30, '30/04/2019': 28, '31/05/2019': 25, '30/06/2019': 35}

dates = list(total_donut.keys())
x_ax = [date[3:5] for date in dates]

y_ax = list(total_donut.values())

plt.plot(x_ax, y_ax)
plt.xlabel('Month')
plt.ylabel('Number of Total Donuts Eaten')
plt.title('Line Plot showing Number of Donuts Eaten over the First 8 months of 2018')
```
